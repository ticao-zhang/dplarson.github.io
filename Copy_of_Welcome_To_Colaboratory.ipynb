With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import pandas as pd
import numpy as np
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.optimizers import Adam
from keras.models import load_model
from keras.callbacks import ModelCheckpoint
```

```
## part 1: reading multiple files
# read training sequence
path_train_UBE = '/content/drive/MyDrive/preprocessed_dataset/trainset_new/UBE_new/user'
path_train_UGE = '/content/drive/MyDrive/preprocessed_dataset/trainset_new/UGE_new/user'
sequences_train = list()
train_target = list()
for i in range(1,151):
    print(i)
    file_path_train_UBE = path_train_UBE + str(i) + '_new.csv'
    file_path_train_UGE = path_train_UGE + str(i) + '_new.csv'
    df_UBE = pd.read_csv(file_path_train_UBE, header=None)
    df_UGE = pd.read_csv(file_path_train_UGE, header=None)
    # print(df.shape[0])
    values_UBE = df_UBE.values
    values_UGE = df_UGE.values
    sequences_train.append(values_UBE)
    train_target.append(0)
    sequences_train.append(values_UGE)
    train_target.append(1)
```

```
# read validation sequence
path_validation_UBE = '/content/drive/MyDrive/preprocessed_dataset/validationset_new/UBE_new/user'
path_validation_UGE = '/content/drive/MyDrive/preprocessed_dataset/validationset_new/UGE_new/user'
sequences_validation = list()
validation_target = list()
for i in range(151,201):
    print(i)
    file_path_validation_UBE = path_validation_UBE + str(i) + '_new.csv'
    file_path_validation_UGE = path_validation_UGE + str(i) + '_new.csv'
    # print(file_path)
    df_UBE = pd.read_csv(file_path_validation_UBE, header=None)
    df_UGE = pd.read_csv(file_path_validation_UGE, header=None)
    # print(df.shape[0])
    values_UBE = df_UBE.values
    values_UGE = df_UGE.values
    sequences_validation.append(values_UBE)
    validation_target.append(0)
    sequences_validation.append(values_UGE)
    validation_target.append(1)
```

```
len_sequences_train = []
for one_seq in sequences_train:
    len_sequences_train.append(len(one_seq))
print(pd.Series(len_sequences_train).quantile(.9))  # 90% length

len_sequences_validation = []
for one_seq in sequences_validation:
    len_sequences_validation.append(len(one_seq))
print(pd.Series(len_sequences_validation).quantile(.9))  # 90% length
```

```


# Padding the sequence with the values in last row to max length
to_pad = math.floor(pd.Series(len_sequences_train).quantile(0.9))
new_seq_train = []
new_seq_validation = []
for one_seq in sequences_train:
    len_one_seq = len(one_seq)
    # last_val = one_seq[-1]
    n = to_pad - len_one_seq
    if n > 0:
        div = n // len_one_seq
        mod = n % len_one_seq
        one_seq_temp = one_seq
        while div > 0:
            one_seq_temp = np.concatenate([one_seq_temp, one_seq])
            div = div - 1
        to_concat = one_seq[0:mod]
        new_one_seq = np.concatenate([one_seq_temp, to_concat])
        # print('Yes')
        # print(len(new_one_seq))

    else:
        new_one_seq = one_seq[0:to_pad]
        # print('No')
        # print(len(new_one_seq))

    new_seq_train.append(new_one_seq)

```

```
final_train_seq = np.stack(new_seq_train)
train = np.array(final_train_seq)
train_target = np.array(train_target)
```

```

for one_seq in sequences_validation:
    len_one_seq = len(one_seq)
    # last_val = one_seq[-1]
    n = to_pad - len_one_seq
    if n > 0:
        # to_concat = np.repeat(one_seq[-1], n).reshape(8, n).transpose()
        # new_one_seq = np.concatenate([one_seq, to_concat])
        div = n//len_one_seq
        mod = n%len_one_seq
        one_seq_temp = one_seq
        while div>0:
          one_seq_temp = np.concatenate([one_seq_temp, one_seq])
          div = div-1
        to_concat = one_seq[0:mod]
        new_one_seq = np.concatenate([one_seq_temp, to_concat])
        print('Yes')
        print(len(new_one_seq))
    else:
        new_one_seq = one_seq[0:to_pad]
    new_seq_validation.append(new_one_seq)
final_validation_seq = np.stack(new_seq_validation)
validation = np.array(final_validation_seq)
validation_target = np.array(validation_target)
```

```
from keras.layers import Dropout
model = Sequential()
model.add(LSTM(3000, input_shape=(to_pad, 8)))
model.add(Dropout(0.5))
# model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

adam = Adam(learning_rate=0.001)
chk = ModelCheckpoint('best_model.pkl', monitor='val_acc', save_best_only=True, mode='max', verbose=1)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
model.summary()
model.fit(train, train_target, epochs=200, batch_size=256, callbacks=[chk], validation_data=(validation,validation_target))

```

```
## loading the model and checking accuracy on test data
# model = load_model('best_model.pkl')
from sklearn.metrics import accuracy_score
validation_preds = model.predict_classes(validation)
accuracy_score(validation_target, validation_preds)
# test_preds = model.predict_classes(test)
# accuracy_score(test_target, test_preds)
```
